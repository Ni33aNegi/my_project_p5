```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('E:\Data analysis\Excel\Customer Churn.csv')
```

```python
df.head(100)
```

```python
def conv(value):
    if value==1:
        return "Yes"
    else:
        return "No"
df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
```

```python
df.head(100)
```

```python
df.info()
```

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ",0)
df["TotalCharges"]=df["TotalCharges"].astype("float")
```

```python
df.isnull().sum().sum() 
```

```python
ax=sns.countplot(x="Churn",data=df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
def conv(value):
    if value=="No":
        return "0"
    else:
        return "1"
df["Churn"]=df["Churn"].apply(conv)
```

```python
df.head(100)
```

```python
plt.figure(figsize=(3,4))
gb=df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"],labels=gb.index,autopct="%0.2f%%")
plt.show()
```

```python
plt.figure(figsize=(3,4))
ax=sns.countplot(data=df,x="gender",hue="Churn")
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
plt.figure(figsize=(3,4))
ax=sns.countplot(data=df,x="SeniorCitizen")
plt.title("Churn by Senior Citizen")
for container in ax.containers:
    ax.bar_label(container)
    plt.legend(title="Churn", loc="upper right")
plt.show()
```

```python
# 1. Crosstab (count SeniorCitizen vs Churn)
ct = pd.crosstab(df["SeniorCitizen"], df["Churn"])

# 2. Normalize to percentages row-wise
ct_perc = ct.div(ct.sum(axis=1), axis=0) * 100

# 3. Plot stacked bar chart
ax = ct_perc.plot(kind="bar", stacked=True, figsize=(4,4))

# 4. Add percentage labels
for container in ax.containers:
    ax.bar_label(container, fmt="%.1f%%", label_type="center")

plt.title("Churn % by Senior Citizen")
plt.ylabel("%")
plt.xlabel("Senior Citizen")
# plt.legend(title="Churn")
plt.show()
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# subplot grid size (3 rows x 3 cols for 9 features)
fig, axes = plt.subplots(3, 3, figsize=(12, 12))

# flatten axes for easy iteration
axes = axes.flatten()

# loop through columns and create countplots
for i, col in enumerate(cols):
    sns.countplot(data=df, x=col, ax=axes[i],hue=df["Churn"])
    axes[i].set_title(col)
    axes[i].tick_params(axis='x')  # rotate labels if needed

# adjust layout
plt.tight_layout()
plt.show()
```
